```python
df = pd.read_excel('ranking_dataset_new.xlsx')
df.head()
```

### Описание
**id** - _идентификатор публикации_,

**authour_name** - _имя автора_,

**title** - _название публикации_,

**issn** - _номер ISSN журнала_,

**journal** - _название журнала_,

**year** - _год издания_,

**address** - _полная аффилиация автора_.

```python
org_names = []
for i in range(len(df)):
    names = df['address'].str.split(',')[i][0]
    org_names.append(names)
df['org_name'] = org_names

df['country'] = df['address'].str.extract('([A-Za-z]+)\.')
```
